```python
import pandas as pd

# Load datasets
patients = pd.read_csv("/content/patients.csv")
doctors = pd.read_csv("/content/doctors.csv")
appointments = pd.read_csv("/content/appointments.csv")
billing = pd.read_csv("/content/billing.csv")
treatments = pd.read_csv("/content/treatments.csv")

```

```python
patients.head()
```

```python
patients.info()
```

```python
patients['date_of_birth'] = pd.to_datetime(patients['date_of_birth'])
patients['registration_date'] = pd.to_datetime(patients['registration_date'])
```

```python
patients.isnull().sum()
```

```python
patients.duplicated().sum()
```

```python
appointments.head()
```

```python
appointments.info()
```

```python
appointments['AppointmentDateTime'] = pd.to_datetime(
    appointments['appointment_date'] + ' ' + appointments['appointment_time']
)
```

```python
appointments.isnull().sum()
```

```python
appointments.duplicated().sum()
```

```python
appointments.head()
```

```python
appointments.drop(columns=['appointment_date','appointment_time'])
```

```python
appointments['Hour'] = appointments['AppointmentDateTime'].dt.hour
appointments['Day'] = appointments['AppointmentDateTime'].dt.day_name()
appointments['Month'] = appointments['AppointmentDateTime'].dt.month
appointments['Weekday'] = appointments['AppointmentDateTime'].dt.weekday
```

```python
appointments.head()
```

```python
billing.head()
```

```python
billing.info()
```

```python
billing['bill_date'] = pd.to_datetime(billing['bill_date'])
```

```python
doctors.head()
```

```python
doctors.info()
```

```python
treatments.head()
```

```python
treatments.info()
```

```python
treatments['treatment_date'] = pd.to_datetime(treatments['treatment_date'])
```

```python
df = appointments.merge(patients,on='patient_id',how='left')
```

```python
df = df.merge(doctors, on='doctor_id',how='left')
```

```python
df = df.merge(billing,on='patient_id',how='left')
```

```python
df = df.merge(treatment,on='appointment_id',how = 'left')
```

```python
df.head()
```

```python
treatment_agg = treatments.groupby('appointment_id').agg({
    'cost': 'sum',
    'treatment_type': 'count'
}).reset_index()

treatment_agg.rename(columns={
    'cost': 'TotalTreatmentCost',
    'treatment_type': 'NumTreatments'
}, inplace=True)

df = df.merge(treatment_agg, on='appointment_id', how='left')
```

```python
df
```

```python
print(df.shape)
print(df.head())

# Check duplicates after merge
print(df.duplicated(subset=['appointment_id']).sum())
```

```python
df.columns
```

```python
df.rename(columns={
    'first_name_x': 'PatientFirstName',
    'last_name_x': 'PatientLastName',
    'first_name_y': 'DoctorFirstName',
    'last_name_y': 'DoctorLastName',
    'email_x': 'PatientEmail',
    'email_y': 'DoctorEmail'
}, inplace=True)
```

```python
df.drop(columns=[
    'treatment_id_x',
    'treatment_id_y'
], inplace=True)
```

```python
df.info()
```

```python
df['Revenue'] = df['amount'].fillna(0)
```

```python
df['TimeSlot'] = pd.cut(
    df['Hour'],
    bins=[0, 6, 12, 18, 24],
    labels=['Night', 'Morning', 'Afternoon', 'Evening']
)
```

```python
df['DoctorName'] = df['DoctorFirstName'] + ' ' + df['DoctorLastName']
```

```python
# Check duplicates
print(df.duplicated(subset=['appointment_id']).sum())

# Check nulls
print(df.isnull().sum())

# Check revenue distribution
print(df['Revenue'].describe())
```

```python
df['PatientCount'] = 1
```

```python
dept_summary = df.groupby('specialization').agg({
    'PatientCount': 'sum',
    'Revenue': 'sum'
}).reset_index()
```

```python
dept_summary
```

```python
doctor_summary = df.groupby('DoctorName').agg({
    'PatientCount': 'sum',
    'Revenue': 'sum'
}).reset_index()
```

```python
doctor_summary
```

```python
time_summary = df.groupby('Hour')['PatientCount'].sum().reset_index()
```

```python
time_summary
```

```python
day_summary = df.groupby('Day')['PatientCount'].sum().reset_index()
```

```python
day_summary
```

```python
treatment_summary = df.groupby('treatment_type').agg({
    'PatientCount': 'sum',
    'Revenue': 'sum'
}).reset_index()
```

```python
df.to_csv("hospital_final.csv", index=False)

dept_summary.to_csv("dept_summary.csv", index=False)
doctor_summary.to_csv("doctor_summary.csv", index=False)
time_summary.to_csv("time_summary.csv", index=False)
day_summary.to_csv("day_summary.csv", index=False)
treatment_summary.to_csv("treatment_summary.csv", index=False)
```
